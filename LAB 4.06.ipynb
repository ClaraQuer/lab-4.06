```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as st
```

```python
customer_df = pd.read_csv('.\\files_for_lab\\we_fn_use_c_marketing_customer_value_analysis.csv')
customer_df
```

```python
customer_df.columns = [i.lower().replace(' ', '_') for i in customer_df.columns]
```

### Get the numerical variables from our dataset.

```python
customer_df_num = customer_df.select_dtypes(np.number)
customer_df_num.dtypes
```

### Check using a distribution plot if the variables fit the theoretical normal or exponential distribution.

```python
def plotter_d(df):
    for col in df:
        sns.displot(df[col])
        plt.show()
```

```python
plotter_d(customer_df_num)
```

The great majority of the plots look like an exponential distribution, but none of them look like a normal one. 

### Check if any of the transformations (log-transform, etc.) we have seen up to this point changes the result.

```python
customer_df_num_log = customer_df_num.apply(np.log)
customer_df_num_log = customer_df_num.replace([np.inf, -np.inf], 0)
customer_df_num_log
```

```python
plotter_d(customer_df_num_log)
```

With the log transformation, they look a little bit more like standard distributions but it still doesn't resemble the actual shape.
